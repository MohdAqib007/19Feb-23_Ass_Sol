```python
#Q2. Create a simple Flask application to display ‘Hello World!!’. Attach the screenshot of the output in Jupyter Notebook.
#Ans-

from flask import Flask

app = Flask(__name__)

@app.route("/")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q3. What is App routing in Flask? Why do we use app routes?
#Ans-

'''In Flask, app routing refers to the process of mapping URLs to specific functions or views in your web application. 
This allows users to navigate your website by typing in a URL or clicking on a link, which then triggers a specific function or view to be executed on the server.

App routes are used in Flask to define the different pages or views of a web application, as well as the actions that users can perform on those pages. 
They are defined using the @app.route decorator, which maps a URL pattern to a Python function.
```

```python
''''Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”
route to show the following details:
Company Name: ABC Corporation
Location: India
Contact Detail: 999-999-9999''''

#Ans-

from flask import Flask

app = Flask(__name__)

@app.route("/welcome")
def welcome():
    return "<h1>Welcome to ABC Corporation</h1>"

@app.route("/copr_details")
def copr_details():
    return "<h1>Company Name: ABC Corporation<br>Location: India<br>Contact Detail: 999-999-9999</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0")
```

```python
#Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

#Ans-

from flask import Flask, url_for

app = Flask(__name__)

@app.route('/')
def home():
    return 'Welcome to the home page!'

@app.route('/user/<name>')
def user(name):
    return f'Hello, {name}!'

@app.route('/login')
def login():
    return 'Please log in to continue'

with app.test_request_context():
    print(url_for('home'))   # Output: '/'
    print(url_for('user', name='John'))   # Output: '/user/John'
    print(url_for('login'))  # Output: '/login'



'''In this example, we create three routes: /, /user/<name>, and /login. 
The home() function returns a welcome message, the user(name) function returns a personalized greeting, and the login() function returns a message prompting the user to log in.

We then use the url_for() function to generate URLs for these endpoints. In the test_request_context() block, we call the url_for() function with the endpoint name and any variable parts that are required. 
The function returns the URL for the given endpoint, which we then print to the console.'''
```
